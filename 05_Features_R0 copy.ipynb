```python
import numpy as np
from numba import njit

@njit(fastmath=True)
def pearson_correlation(xk, x1):
    """
    Calcula a correlao de Pearson entre dois vetores xk e x1 
    usando Numba para alta performance.
    """
    n = len(xk)
    
    # Verifica se os tamanhos so compatveis
    if n != len(x1):
        raise ValueError("Os arrays devem ter o mesmo tamanho.")

    # 1. Clculo das mdias (bar_xk e bar_x1 na frmula)
    mean_k = 0.0
    mean_1 = 0.0
    for i in range(n):
        mean_k += xk[i]
        mean_1 += x1[i]
    mean_k /= n
    mean_1 /= n

    # 2. Clculo do Numerador e Denominador
    numerator = 0.0
    sum_sq_diff_k = 0.0
    sum_sq_diff_1 = 0.0

    for i in range(n):
        diff_k = xk[i] - mean_k
        diff_1 = x1[i] - mean_1
        
        # Somatrio do produto das diferenas (Numerador)
        numerator += diff_k * diff_1
        
        # Somatrio dos quadrados das diferenas (partes do Denominador)
        sum_sq_diff_k += diff_k ** 2
        sum_sq_diff_1 += diff_1 ** 2

    # O denominador  a raiz do produto das somas dos quadrados
    # Frmula: sqrt( sum((xk-mean_k)^2) * sum((x1-mean_1)^2) )
    denominator = np.sqrt(sum_sq_diff_k * sum_sq_diff_1)

    if denominator == 0:
        return 0.0  # Evita diviso por zero se a varincia for nula

    return numerator / denominator
```

```python
import numpy as np
import pandas as pd
import os
import re
import matplotlib.pyplot as plt
from numba import njit

# 1. Funo Numba Otimizada (Semi-Vectorizada)
@njit(fastmath=True)
def fast_pearson_precalc(xk, x1_centered, x1_norm):
    """
    Calcula Pearson assumindo que x1 j est centralizado e sua norma calculada.
    Isso corta as operaes matemticas pela metade.
    """
    n = len(xk)
    
    # Passo 1: Calcular mdia de xk
    mean_k = 0.0
    for i in range(n):
        mean_k += xk[i]
    mean_k /= n

    # Passo 2: Calcular numerador (produto escalar) e norma de xk
    numerator = 0.0
    sum_sq_diff_k = 0.0
    
    for i in range(n):
        diff_k = xk[i] - mean_k
        
        # O x1_centered[i] j foi calculado fora do loop!
        numerator += diff_k * x1_centered[i]
        sum_sq_diff_k += diff_k ** 2

    # Passo 3: Finalizar
    denom_k = np.sqrt(sum_sq_diff_k)
    
    # Evita diviso por zero
    if denom_k == 0 or x1_norm == 0:
        return 0.0
        
    return numerator / (denom_k * x1_norm)

# --- Configurao do Diretrio ---
dir_path = '03_FTRSY_WND/' # Renomeei para evitar conflito com built-in dir()
out = '03_VMDY_FFT_CP/'
brngs = os.listdir(dir_path)
brngs.sort(key=lambda x: int(re.search(r'\d+', x).group()))

# Loop Principal
for brng in brngs:
    folder_path = os.path.join(dir_path, brng)
    
    # Filtro especfico do seu cdigo
    #if not brng.endswith('Bearing1_2'):
    #    continue

    # Leitura do arquivo de referncia (FEITO APENAS UMA VEZ POR PASTA)
    df1 = pd.read_csv(os.path.join(folder_path, '1.csv'))
    if dir_path.endswith('FFT/'): x1 = df1['Mode_1'].values.astype(np.float64) # Garantir float64 para numba
    else: x1 = df1['Entropia'].values.astype(np.float64) # Garantir float64 para numba

    # --- PR-CLCULO DO X1 (Otimizao Chave) ---
    x1_mean = np.mean(x1)
    x1_centered = x1 - x1_mean
    # Norma Euclidiana (parte do denominador da frmula de Pearson)
    x1_norm = np.sqrt(np.sum(x1_centered**2))
    
    # Listar amostras
    samples = os.listdir(folder_path)
    samples = [s for s in samples if s.endswith('.csv')] # Garante que s pega csv
    samples.sort(key=lambda x: int(re.search(r'\d+', x).group()))
    
    num_samples = len(samples)
    
    # --- OTIMIZAO DE MEMRIA ---
    # Alocar array vazio ao invs de usar np.append
    corr_results = np.zeros(num_samples)
    
    # Loop das amostras
    for idx, sample in enumerate(samples):
        # I/O ainda  o gargalo aqui. 
        # Se for muito lento, veja a dica sobre Joblib abaixo.
        df = pd.read_csv(os.path.join(folder_path, sample))
        if dir_path.endswith('FFT/'): xk = df['Mode_1'].values.astype(np.float64)
        else: xk = df['Entropia'].values.astype(np.float64) # Garantir float64 para numba

        # Chama a funo otimizada
        # Note que passamos x1_centered e x1_norm, no o x1 original
        if len(xk) == len(x1_centered):
            corr_results[idx] = fast_pearson_precalc(xk, x1_centered, x1_norm)
        else:
            # Caso tamanhos difiram, trata erro ou ignora
            corr_results[idx] = 0.0
    df = pd.DataFrame()
    df['RS'] = np.abs(corr_results)
    #df.to_csv(out+brng+'.csv',index=False)  
    break
```

```python
import numpy as np
import pandas as pd
import os
import re
from numba import njit

# 1. Funo Numba (Mantida idntica pois  eficiente para vetores 1D)
@njit(fastmath=True)
def fast_pearson_precalc(xk, x1_centered, x1_norm):
    """
    Calcula Pearson entre vetor xk e referncia x1 (j pr-processada).
    """
    n = len(xk)
    
    # Passo 1: Calcular mdia de xk
    mean_k = 0.0
    for i in range(n):
        mean_k += xk[i]
    mean_k /= n

    # Passo 2: Calcular numerador e soma dos quadrados de xk
    numerator = 0.0
    sum_sq_diff_k = 0.0
    
    for i in range(n):
        diff_k = xk[i] - mean_k
        numerator += diff_k * x1_centered[i]
        sum_sq_diff_k += diff_k ** 2

    # Passo 3: Finalizar
    denom_k = np.sqrt(sum_sq_diff_k)
    
    if denom_k == 0 or x1_norm == 0:
        return 0.0
        
    return numerator / (denom_k * x1_norm)

# --- Configurao ---
dir_path = '02_FTRSX_WND/'
out = '02_FTRSX_WND_CP/'

# Garante que o diretrio de sada existe
os.makedirs(out, exist_ok=True)

brngs = os.listdir(dir_path)
# Filtra apenas diretrios para evitar erro com arquivos soltos
brngs = [b for b in brngs if os.path.isdir(os.path.join(dir_path, b))]
brngs.sort(key=lambda x: int(re.search(r'\d+', x).group()) if re.search(r'\d+', x) else 0)

# Loop Principal (Por Pasta/Bearing)
for brng in brngs:
    folder_path = os.path.join(dir_path, brng)
    
    # 1. Ler arquivo de referncia
    ref_file_path = os.path.join(folder_path, '1.csv')
    if not os.path.exists(ref_file_path):
        print(f"Arquivo de referncia no encontrado em {brng}. Pulando.")
        continue
        
    df1 = pd.read_csv(ref_file_path)
    
    # Identificar colunas numricas automaticamente
    # Isso evita erros se houver colunas de texto/index
    cols = df1.select_dtypes(include=[np.number]).columns.tolist()
    
    # Estrutura para armazenar os dados pr-calculados de CADA coluna
    # Dicionrio: {'NomeColuna': (array_centralizado, float_norma)}
    ref_stats = {}
    
    for col in cols:
        x1 = df1[col].values.astype(np.float64)
        x1_mean = np.mean(x1)
        x1_centered = x1 - x1_mean
        x1_norm = np.sqrt(np.sum(x1_centered**2))
        
        # Armazena na memria para uso rpido no loop
        ref_stats[col] = (x1_centered, x1_norm)

    # 2. Preparar lista de amostras
    samples = [s for s in os.listdir(folder_path) if s.endswith('.csv')]
    samples.sort(key=lambda x: int(re.search(r'\d+', x).group()) if re.search(r'\d+', x) else 0)
    num_samples = len(samples)

    # 3. Alocar matriz de resultados
    # Shape: (Linhas=Amostras, Colunas=Colunas do CSV)
    # Usamos um dicionrio de arrays numpy para preencher coluna por coluna
    results_data = {col: np.zeros(num_samples, dtype=np.float64) for col in cols}
    
    print(f"Processando {brng} com {len(cols)} colunas e {num_samples} arquivos...")

    # Loop das amostras
    for idx, sample in enumerate(samples):
        df = pd.read_csv(os.path.join(folder_path, sample))
        
        # Itera sobre cada coluna que existe no arquivo de referncia
        for col in cols:
            # Verifica se a coluna existe no arquivo atual (segurana)
            if col in df.columns:
                xk = df[col].values.astype(np.float64)
                
                # Recupera os dados pr-calculados da referncia
                x1_centered, x1_norm = ref_stats[col]
                
                # Verifica compatibilidade de tamanho
                if len(xk) == len(x1_centered):
                    corr = fast_pearson_precalc(xk, x1_centered, x1_norm)
                    results_data[col][idx] = abs(corr) # Aplica ABS conforme seu cdigo original
                else:
                    results_data[col][idx] = 0.0
            else:
                results_data[col][idx] = 0.0

    # 4. Criar DataFrame Final e Salvar
    df_results = pd.DataFrame(results_data)
    break
    
    # Salva com o mesmo nome da pasta
    #output_file = os.path.join(out, brng + '.csv')
    #df_results.to_csv(output_file, index=False)
    #print(f"Salvo: {output_file}")
```

```python
import numpy as np
import pandas as pd
import numba as nb

# ==============================================================================
# 1. FUNES AUXILIARES OTIMIZADAS COM NUMBA
# ==============================================================================

@nb.njit(fastmath=True)
def _calc_skew_kurt(x, mean, std, n):
    """Calcula Skewness e Kurtosis (Fisher) em uma nica passada."""
    if std == 0:
        return 0.0, 0.0
    
    sum_pow3 = 0.0
    sum_pow4 = 0.0
    
    for i in range(n):
        val = (x[i] - mean) / std
        sum_pow3 += val ** 3
        sum_pow4 += val ** 4
        
    skew = sum_pow3 / n
    kurt = (sum_pow4 / n) - 3.0  # Fisher Kurtosis (normal = 0)
    return skew, kurt

@nb.njit(fastmath=True)
def _calc_entropy_spectral(power_spectrum):
    """Calcula Entropia de Shannon normalizada sobre um espectro de energia."""
    total_energy = np.sum(power_spectrum)
    if total_energy == 0:
        return 0.0
        
    entropy = 0.0
    for i in range(len(power_spectrum)):
        p = power_spectrum[i] / total_energy
        if p > 0:
            entropy -= p * np.log(p)
    return entropy

@nb.njit(fastmath=True)
def _calc_entropy_histogram(x, bins=10):
    """Calcula entropia simples baseada em histograma para sries temporais."""
    n = len(x)
    if n == 0: return 0.0
    
    mn = np.min(x)
    mx = np.max(x)
    if mn == mx: return 0.0
    
    # Histograma manual
    hist = np.zeros(bins, dtype=np.float64)
    width = (mx - mn) / bins
    
    for i in range(n):
        idx = int((x[i] - mn) / width)
        if idx >= bins: idx = bins - 1
        hist[idx] += 1
    
    # Entropia
    entropy = 0.0
    for i in range(bins):
        p = hist[i] / n
        if p > 0:
            entropy -= p * np.log(p)
    return entropy

# ==============================================================================
# 2. EXTRAO DE FEATURES (TIME & FREQ)
# ==============================================================================

@nb.njit(fastmath=True)
def get_time_features_numba(vec):
    n = len(vec)
    
    # Pr-clculos bsicos
    mean_val = np.mean(vec)
    abs_vec = np.abs(vec)
    mean_abs = np.mean(abs_vec)
    
    # Varincia e Std
    var_val = np.var(vec)
    std_val = np.sqrt(var_val)
    
    # RMS
    # mean(square)  equivalente a var + mean^2, mas calculamos direto para preciso
    rms = np.sqrt(np.mean(vec**2))
    
    # Stats avanados (substitui scipy)
    skw, krt = _calc_skew_kurt(vec, mean_val, std_val, n)
    ntrpy = _calc_entropy_histogram(vec)
    
    max_val = np.max(vec)
    min_val = np.min(vec)
    ptp = max_val - min_val
    
    crest = max_val / rms if rms != 0 else 0.0
    ff = rms / mean_abs if mean_abs != 0 else 0.0
    
    # Zero Crossing Rate (Implementao manual rpida)
    zcr_count = 0
    for i in range(n - 1):
        if (vec[i] >= 0 and vec[i+1] < 0) or (vec[i] < 0 and vec[i+1] >= 0):
            zcr_count += 1
            
    # Retorna array fixo (mais rpido que lista)
    # Ordem: [Mean, MeanAbs, Var, Std, Skew, Kurt, Entropy, RMS, Max, PTP, Crest, ZCR, FormFactor]
    return np.array([mean_val, mean_abs, var_val, std_val, skw, krt, ntrpy, rms, 
                     max_val, ptp, crest, float(zcr_count), ff])

@nb.njit(fastmath=True)
def get_freq_features_numba(vec, fs):
    n = len(vec)
    
    # --- CORREO DE COMPATIBILIDADE (FIX) ---
    # Usamos objmode para contornar o erro de "Unknown attribute fft"
    # Isso fora o Python a executar o FFT, j que o Numba falhou em linkar nativamente.
    # Definimos que a sada 'fft_complex' ser um array de complexos (complex128[:])
    with nb.objmode(fft_complex='complex128[:]'):
        fft_complex = np.fft.fft(vec)
    
    # ----------------------------------------

    # Pegamos apenas a metade positiva do espectro
    limit_idx = n // 2 + 1
    
    # Slice manual
    fft_mag = np.abs(fft_complex[:limit_idx])
    
    # Criar vetor de frequncias manualmente
    freqs = np.empty(limit_idx, dtype=np.float64)
    for i in range(limit_idx):
        freqs[i] = i * (fs / n)
    
    # Filtragem de frequncia (1 a 1500 Hz)
    idx_start = 0
    idx_end = limit_idx 
    
    found_start = False
    for i in range(limit_idx):
        if not found_start and freqs[i] >= 1:
            idx_start = i
            found_start = True
        if freqs[i] > 1500:
            idx_end = i
            break
            
    # Proteo: Se o filtro resultar em vazio
    if idx_start >= idx_end:
        return np.zeros(15, dtype=np.float64)

    # Fatiamento (Slicing)
    fft_sel = fft_mag[idx_start:idx_end]
    freq_sel = freqs[idx_start:idx_end]
    
    # Potncia (Square)
    power_spec = fft_sel ** 2 
    
    # Estatsticas no Domnio da Frequncia
    energy = np.sum(power_spec)
    
    if energy == 0:
        return np.zeros(15, dtype=np.float64)

    mean_en = np.mean(power_spec)
    var_en = np.var(power_spec)
    std_en = np.sqrt(var_en)
    
    rms_en = np.sqrt(np.mean(power_spec**2)) 
    
    # Skew/Kurt da Energia (usando a funo auxiliar compilada anteriormente)
    skw_en, krt_en = _calc_skew_kurt(power_spec, mean_en, std_en, len(power_spec))
    
    # Entropia (usando a funo auxiliar compilada anteriormente)
    ntrpy_en = _calc_entropy_spectral(power_spec)
    
    # Density
    total_energy_orig = np.sum(fft_mag**2)
    energy_density = energy / total_energy_orig if total_energy_orig > 0 else 0.0
    
    # Parmetros Espectrais
    argmax_idx = np.argmax(power_spec)
    peak_freq = freq_sel[argmax_idx]
    max_amp_fft = fft_sel[argmax_idx]
    
    # Centroid e Bandwidth
    num_cent = np.sum(freq_sel * power_spec)
    centroid = num_cent / energy
    
    num_bw = np.sum(((freq_sel - centroid)**2) * power_spec)
    bandwidth = np.sqrt(num_bw / energy)
    
    # Flatness
    mean_arith = mean_en
    log_sum = 0.0
    for val in power_spec:
        log_sum += np.log(val + 1e-16) # Epsilon para evitar log(0)
        
    mean_geom = np.exp(log_sum / len(power_spec))
    flatness = mean_geom / mean_arith if mean_arith > 0 else 0.0

    # Rolloff (85% da energia)
    target_en = 0.85 * energy
    cumsum = 0.0
    rolloff = 0.0
    for i in range(len(power_spec)):
        cumsum += power_spec[i]
        if cumsum >= target_en:
            rolloff = freq_sel[i]
            break
            
    peak_to_rms = max_amp_fft / rms_en if rms_en > 0 else 0.0

    return np.array([energy, mean_en, var_en, std_en, skw_en, ntrpy_en, krt_en, 
                     energy_density, rms_en, peak_freq, centroid, bandwidth, 
                     flatness, rolloff, peak_to_rms])
# ==============================================================================
# 3. ROTINA DE JANELA DESLIZANTE (SLIDING WINDOW)
# ==============================================================================

@nb.njit(parallel=True) # Paraleliza o loop das 100 janelas
def extract_sliding_features(full_vec, fs, window_size, num_steps):
    """
    Extrai features para 'num_steps' janelas consecutivas.
    """
    # 13 features de tempo + 15 features de frequncia = 28 colunas
    num_features_time = 13
    num_features_freq = 15
    total_cols = num_features_time + num_features_freq
    
    results = np.zeros((num_steps, total_cols))
    
    # Loop principal (pode ser paralelizado)
    for i in nb.prange(num_steps):
        # Define a janela: v_i = vec[i : i + window_size]
        # Nota: assumindo passo de 1 amostra conforme sua descrio (vec_0...vec_n, vec_1...vec_n+1)
        start_idx = i
        end_idx = i + window_size
        
        # Proteo de ndice
        if end_idx > len(full_vec):
            break
            
        window = full_vec[start_idx : end_idx]
        
        # Extrai Time Features
        t_feats = get_time_features_numba(window)
        
        # Extrai Freq Features
        f_feats = get_freq_features_numba(window, fs)
        
        # Preenche matriz de resultados
        for j in range(num_features_time):
            results[i, j] = t_feats[j]
            
        for k in range(num_features_freq):
            results[i, num_features_time + k] = f_feats[k]
            
    return results

# ==============================================================================
# 4. EXEMPLO DE USO (WRAPPER PYTHON)
# ==============================================================================

def process_signal_sliding(vec, fs, num_windows=100):
    """Funo Python amigvel para chamar o motor Numba."""
    
    # Definir tamanho da janela
    # Se queremos 100 janelas com passo 1, a janela deve ser len(vec) - 99
    # Ou o usurio define um tamanho fixo. Vou assumir o mximo possvel.
    window_size = len(vec) - num_windows + 1
    
    if window_size <= 0:
        raise ValueError("Sinal muito curto para o nmero de janelas solicitado.")

    # Garante float64 e array contguo para performance mxima
    vec_nb = np.ascontiguousarray(vec, dtype=np.float64)
    
    # Chama Numba
    matrix_results = extract_sliding_features(vec_nb, fs, window_size, num_windows)
    
    # Criar DataFrame
    cols_time = ['Mdia', 'Mdia Absoluta', 'Varincia', 'Desvio Padro', 
                 'Assimetria', 'Kurtose', 'Entropia', 'RMS', 'Mximo', 'Pico-a-pico',
                 'Fator de Cresta', 'ZCR', 'Fator de Forma']
                 
    cols_freq = ['Energia', 'Energia Mdia', 'Varincia Energia', 'Std Energia', 
                 'Assimetria Energia', 'Entropia Energia', 'Kurtose Energia', 
                 'Densidade Energia', 'RMS Energia', 'Freq Pico', 'Centroide', 
                 'Largura Banda', 'Planicidade', 'Rolloff', 'Pico/RMS']
                 
    df = pd.DataFrame(matrix_results, columns=cols_time + cols_freq)
    return df

path = r'02_VMDX\Bearing1_1\1.csv'
df = pd.read_csv(path)
sig = df['IMF1']
df = process_signal_sliding(sig, fs, num_windows=100)
```

```python
import numpy as np
import pandas as pd
import numba as nb

# ... (MANTENHA AS FUNES _calc_*, get_time_features_numba E get_freq_features_numba AQUI) ...

# ==============================================================================
# NOVA FUNO DE EXTRAO (DIVISO CONSECUTIVA)
# ==============================================================================

@nb.njit(parallel=True)
def extract_split_features(full_vec, fs, num_splits):
    """
    Divide o vetor em 'num_splits' partes iguais e consecutivas 
    e extrai as features de cada parte.
    """
    total_len = len(full_vec)
    
    # Define o tamanho de cada bloco (diviso inteira)
    # Ex: 1000 pontos / 100 splits = 10 pontos por bloco
    block_size = total_len // num_splits
    
    if block_size < 2:
        # Retorna matriz de zeros se o bloco for pequeno demais para calcular features
        return np.zeros((num_splits, 28)) # 13 time + 15 freq

    # Aloca matriz de resultados
    # 13 features tempo + 15 features frequncia = 28 colunas
    results = np.zeros((num_splits, 28))
    
    # Loop Paralelo (Cada iterao trata um bloco independente)
    for i in nb.prange(num_splits):
        # Clculo dos ndices (v1, v2, ... v100)
        start_idx = i * block_size
        end_idx = start_idx + block_size
        
        # Extrai o segmento do sinal original
        segment = full_vec[start_idx : end_idx]
        
        # --- Chama as funes de features j compiladas ---
        t_feats = get_time_features_numba(segment)
        f_feats = get_freq_features_numba(segment, fs)
        
        # Preenche a matriz (Time Features: colunas 0 a 12)
        for j in range(13):
            results[i, j] = t_feats[j]
            
        # Preenche a matriz (Freq Features: colunas 13 a 27)
        for k in range(15):
            results[i, 13 + k] = f_feats[k]
            
    return results

# ==============================================================================
# NOVO WRAPPER PYTHON
# ==============================================================================

def process_signal_splits(vec, fs, num_splits=100):
    """
    Wrapper para dividir o sinal em 100 partes e gerar o DataFrame.
    """
    # Verificaes bsicas
    if len(vec) < num_splits:
        raise ValueError(f"O sinal  muito curto ({len(vec)}) para ser dividido em {num_splits} partes.")
    
    # Garante array contguo e tipo correto para o Numba
    vec_nb = np.ascontiguousarray(vec, dtype=np.float64)
    
    # Chama a funo Numba paralelizada
    matrix_results = extract_split_features(vec_nb, fs, num_splits)
    
    # Definio dos nomes das colunas
    cols_time = ['Mdia', 'Mdia Absoluta', 'Varincia', 'Desvio Padro', 
                 'Assimetria', 'Kurtose', 'Entropia', 'RMS', 'Mximo', 'Pico-a-pico',
                 'Fator de Cresta', 'ZCR', 'Fator de Forma']
                 
    cols_freq = ['Energia', 'Energia Mdia', 'Varincia Energia', 'Std Energia', 
                 'Assimetria Energia', 'Entropia Energia', 'Kurtose Energia', 
                 'Densidade Energia', 'RMS Energia', 'Freq Pico', 'Centroide', 
                 'Largura Banda', 'Planicidade', 'Rolloff', 'Pico/RMS']
    
    # Criao do DataFrame
    # O ndice agora representa o nmero do bloco (0 a 99)
    df = pd.DataFrame(matrix_results, columns=cols_time + cols_freq)
    
    # Opcional: Adicionar coluna indicando o tempo inicial de cada bloco
    # total_duration = len(vec) / fs
    # block_duration = total_duration / num_splits
    # df['Tempo_Inicio'] = np.linspace(0, total_duration - block_duration, num_splits)
    
    return df

path = r'02_VMDX\Bearing1_1\1.csv'
df = pd.read_csv(path)
sig = df['IMF1']
df = process_signal_splits(sig, fs, num_splits=100)
```
